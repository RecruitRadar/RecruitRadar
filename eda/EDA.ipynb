### 회사 주소

```python
temp_df = df.copy()
```

```python
def clean_location_sido(location):
    if location:
        location = location.split()
        if location[0].startswith('주소'):
            location = location[1][:2]
        else:
            location = location[0][:2]
        return location
    else:
        None
```

```python
temp_df['sido_location'] = temp_df['location'].apply(lambda x: clean_location_sido(x))
```

```python
grouped_df_location = temp_df.groupby('sido_location').size().reset_index(name='count')

# 그래프 그리기
plt.figure(figsize=(10, 20))
ax = sns.barplot(data=grouped_df_location, y='sido_location', x='count')
plt.title('Grouped Data Count by Location (Sido)')
plt.xlabel('채용공고 수')
plt.ylabel('지역')
plt.show()
```

```python
def clean_location_gugun(location):
    if location:
        location = location.split()
        if location[0].startswith('주소'):
            location = location[2][:2]
        else:
            if len(location) == 1:
                return None
            location = location[1][:2]
        return location
    else:
        None
```

```python
temp_df['gugun_location'] = temp_df['location'].apply(lambda x: clean_location_gugun(x))
```

```python
grouped_df_location_gugun = temp_df.groupby('gugun_location').size().reset_index(name='count')

# 그래프 그리기
plt.figure(figsize=(10, 30))
ax = sns.barplot(data=grouped_df_location_gugun, y='gugun_location', x='count')
plt.title('Grouped Data Count by Location (Gugun)')
plt.xlabel('채용공고 수')
plt.ylabel('지역')
plt.show()
```

```python
from kiwipiepy import Kiwi
```

```python
kiwi = Kiwi()
```

```python
def extract_noun(text):
    result = kiwi.tokenize(text)
    for token in result:
        if token.tag in ['NNG', 'NNP']:
            yield token.form
```

```python
df['cleaned_required'] = df['required'].apply(lambda x: list(extract_noun(str(x))))
df['cleaned_primary_responsibility'] = df['primary_responsibility'].apply(lambda x: list(extract_noun(x)))
df['cleaned_preferred'] = df['preferred'].apply(lambda x: list(extract_noun(x)))
df['cleaned_welfare'] = df['welfare'].apply(lambda x: list(extract_noun(x)))
df['cleaned_company_description'] = df['company_description'].apply(lambda x: list(extract_noun(x)))
```
